```python
#Import Packages

import pandas as pd
import numpy as np
%matplotlib inline
from urllib.request import urlopen #open url
from bs4 import BeautifulSoup
```

```python
#specify the URL containing the dataset and pass it to urlopen() to get the html of the page.

url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M" #assign the URL
html = urlopen(url) #get the html of the page
```

```python
soup = BeautifulSoup(html,'lxml') #create a beauifulsoup object
type(soup)
```

```python
#Get title
title = soup.title
print(title)
```

```python
#Print the text
text = soup.get_text()
#print(text)
```

```python
all_links = soup.find_all("a")
#for link in all_links:
#    print(link.get("href"))
```

```python
#find all row of the table  with find_all < tr > for table rows
rows = soup.find_all("tr")

print(rows[:10])
```

```python
#get table row

for row in rows:
    row_td = row.find_all("td")
print(row_td)
type(row_td)
```

```python
str_cells = str(row_td) #converts value into a string
cleantext = BeautifulSoup(str_cells, "lxml").get_text() #extract the text
print(cleantext)
```

```python
#the characters inside the < td > html tags and replace them with an empty string for each table row.
import re

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>') #match every character inside the brackets.
    clean2 = (re.sub(clean, '',str_cells)) #find all substring where the regular expression is matched.
    list_rows.append(clean2)
print(clean2,'\n',list_rows)
type(clean2)
```

```python
df = pd.DataFrame(list_rows) #conver to pandas DataFrame
df.head(10)
```

#### Select Borough with "Toronto" in their name

```python
tor = df12[df12["Borough"].str.contains("Toronto")]
tor["Borough"] =tor["Borough"].str.strip()
```

```python
!conda install -c conda-forge geopy --yes  
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans

!conda install -c conda-forge folium=0.5.0 --yes  
import folium # map rendering library
```

```python
address = 'Toronto, CA'

geolocator = Nominatim(user_agent="tor_ex")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

#### Cluster Neighbourhood

```python
k = len(tor["Borough"].unique())
tor_X =tor.drop(["Postcode","Neighbourhood","Borough"], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=k , random_state=0).fit(tor_X)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
cluster = kmeans.labels_ 
len(cluster)
```

```python
tor['Cluster'] = cluster
```

```python
tor
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(tor['Latitude'],tor['Longitude'], tor['Neighbourhood'], tor['Cluster']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
